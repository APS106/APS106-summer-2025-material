```python
def is_proper_name(name):
    """
    (str) -> bool
    
    Determines if the given string is a proper name.

    A proper name has each word starting with a capital letter and words separated by spaces.


    """
    # Check if the first character is uppercase
    if not name[0].isupper():
        return False
    
    i = 1
    while i < len(name):
        # Find the next space
        if name[i] == ' ':
            # Check if the next character is uppercase
            if i + 1 < len(name) and not name[i + 1].isupper():
                return False
            # Skip the space
            i += 1
        i += 1
    
    return True

# Example usage
print(is_proper_name("John Doe"))  # Output: True
print(is_proper_name("john doe"))  # Output: False
print(is_proper_name("John doe"))  # Output: False
print(is_proper_name("John"))      # Output: True
print(is_proper_name("John Doe Smith"))  # Output: True
```

```python
# Write a function that evaluates tanh-1(x) from the series
#
# tanh-1(x) = x + x3/3 + x5/5 + x7/7 + x9/9 + ...
#
# where x < 1.0 and x > -1.0.  The function should take x and, epsilon, 
# and return the estimate when the absolute value of the next term is the 
# series is less than epsilon. 


def estimate_tanh_inv(x, epsilon):
    '''(float, int) -> float
    Returns an estimate of tanh^-1 based on the series:
    tanh-1(x) = x + x3/3 + x5/5 + x7/7 + x9/9 + ...
    Stops when the estimate is with next item is the series is less than
    or equal to epsilon
    '''
    tanh_inv = 0
    term = x
    num = 3
    while abs(term) > epsilon:
        #print(term, num)
        tanh_inv += term
        term = x**num/num
        num += 2
        
    return tanh_inv

x = -2
while x < -1.0 or x > 1.0:
    x = float(input("Enter x between -1 and 1: "))
    eps = float(input("Enter epsilon: "))

tanh_inv = estimate_tanh_inv(x,eps)
print("tanh-1(", x, ") = ", tanh_inv, sep="")
```

Print the following pattern (**HINT** see the pattern, think of a braid or weaving):

```python
# Define the initial strands
strand1 = "A" #or red
strand2 = "B" #or blue
strand3 = "C" #or green
strand4 = "D" #or orange

# Number of times to repeat the braiding process
repeat_times = 10

# Initialize the braid pattern
braid_pattern = ""

# Braiding process using a while loop
counter = 0
while counter < repeat_times:
    # Move the characters to form the braid
    braid_pattern += strand1 + strand2 + strand3 + strand4 + '\n'
    # Simulate the braid by swapping the positions
    temp = strand1
    strand1 = strand2
    strand2 = strand3
    strand3 = strand4
    strand4 = temp
    
    counter += 1 #one line complete

# Display the final braided pattern
print("Final Braid Pattern:\n", braid_pattern, sep='')
```
