```python
#Question 3a

'''



'''


def validate_email(address):
    """
    (str) -> bool

    Returns True if the address is in valid email address format.
    Returns False otherwise.
    
    A valid email address must contain an '@' symbol and a period ('.') after the '@'.
    The period must not come immediately after the '@' and must not be the last character.
    """
    
    if not ('@' in address):
        return False

    at_index = address.find('@')
    period_index = address[at_index:].find('.')

    # Check if period is found, not immediately after '@', and not the last character
    if period_index == -1 or period_index == at_index + 1 or period_index == len(address) - 1:
        return False

    return True
```

```python
#Question 5a

'''

'''

def factorial(n):
    """
    (int) -> int
    Computes the factorial of n (n!)
    
    Example: 
    factorial(3) is equivalent to 3! = 3*2*1 = 6
    factorial(0) = 1
    """

    if n == 0 or n == 1:
    return 1
    
    result = 1
    for i in range(2, n + 1):
        result *= i
    
    return result
```
