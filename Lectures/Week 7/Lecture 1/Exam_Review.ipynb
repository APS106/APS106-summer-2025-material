```python
def remove_punctuation(sentence):
    '''
    (str) -> str
    Takes a sentence as input and returns the sentence with all punctuations replaced by an empty string (""). 

    Example:
    >>> print(remove_punctuation('Hello, how are you?'))
    Hello how are you
    '''

    punctuations = "!-;:,'.?" # punctuation marks to be removed

    # Write your code here. 
    formatted = str(sentence)
    for char in sentence:
        if char in punctuations:
            formatted = formatted.replace(char, '')  
    return formatted
```

```python
def translate(sentence, language):
    '''
    (str, str) -> str
    Takes as input a sentence to translate and the language of that sentence ("eng" or "span"). 
    Returns the sentence translated into the other language. 
    If the sentence is English sentence, the function must return it in Spanish and vice versa. 
    The dictionaries span_to_eng and eng_to_span contain word mappings between the languages for all words in the English and Spanish dictionaries. 
    A sentence may contain punctuation and capital letters but the dictionaries and the translated sentence only contain lower-case letters.
    Example:
    >>> print(translate('I speak English, my friends!!!!', 'eng'))
    yo hablo ingles mis amigos
    '''

    span_to_eng = {'yo': 'i', 'hablo': 'speak', 'espanol': 'spanish'}
    eng_to_span = {'i': 'yo', 'speak': 'hablo', 'spanish': 'espanol'}

    # Write your code here.  
    sentence = remove_punctuation(sentence)
    
    print('sentence: ', sentence)
    
    words = sentence.lower().split(' ')
    print('words: ', words)
    
    
    translated = ''

    for word in words:
        if language == 'eng':
            translated += eng_to_span[word] + ' '
        else:
            translated += span_to_eng[word] + ' '
    return translated
```

Print the following pattern (**HINT** see the pattern, think of a braid or weaving):

ABCD<br>
BCDA<br>
CDAB<br>
DABC<br>
ABCD<br>
BCDA<br>

```python
# Define the initial strands
strand1 = "A" #or red
strand2 = "B" #or blue
strand3 = "C" #or green
strand4 = "D" #or orange

# Number of times to repeat the braiding process
repeat_times = 10

# Initialize the braid pattern
braid_pattern = ""

# Braiding process using a while loop
counter = 0
while counter < repeat_times:
    # Move the characters to form the braid
    braid_pattern += strand1 + strand2 + strand3 + strand4 + '\n'
    # Simulate the braid by swapping the positions
    temp = strand1
    strand1 = strand2
    strand2 = strand3
    strand3 = strand4
    strand4 = temp
    
    counter += 1 #one line complete

# Display the final braided pattern
print("Final Braid Pattern:\n", braid_pattern, sep='')
```

```python
def invert_dictionary(start_dict):
    '''
    (dict) -> dict
    
    
    start_dict = {1:'1', 2:'2', 3:'3', 'three': '3'}
    
    start_dict[key] = value
    
    invert_dict = {'1':1, '2':2, '3': 3}
    
    inverted_dict = {'1':[1], '2':[2], '3':[3, 'three']}
    '''
    
    inverted_dict = {}
    
    
    for item in start_dict:
        
        #if key already exists
        if start_dict[item] in inverted_dict:
            inverted_dict[start_dict[item]].append(item)
        
        else:
            inverted_dict[start_dict[item]] = [item]
    
    return inverted_dict
```
