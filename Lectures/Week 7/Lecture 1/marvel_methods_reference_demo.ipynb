```python
# String methods
s = "Hello, world! Hello, universe!"

# count
print(s.count("Hello"))

# index
print(s.index("world"))

# find
print(s.find("universe"))

# rfind
print(s.rfind("Hello"))

# upper
print(s.upper())

# lower
print(s.lower())

# isupper
print(s.isupper())

# islower
print(s.islower())

# capitalize
print(s.capitalize())

# replace
print(s.replace("Hello", "Hi"))

# split
print(s.split())

# strip
s2 = "   Hello, world!   "
print(s2.strip())
```

```python
# Tuple methods
t = (1, 2, 3, 1, 2, 1)

# count
print(t.count(1))

# index
print(t.index(3))
```

```python
# List methods
l = [1, 2, 3, 4, 5]

# append
l.append(6)
print(l)

# extend
l.extend([7, 8, 9])
print(l)

# count
print(l.count(2))

# insert
l.insert(0, 0)
print(l)

# remove
l.remove(3)
print(l)

# pop
print(l.pop())
print(l)

# index
print(l.index(4))

# reverse
l.reverse()
print(l)

# sort
l.sort()
print(l)
```

```python
# Set methods
s = {"Iron Man", "Thor", "Hulk"}

# add
s.add("Spider-Man")
print(s)

# remove
s.remove("Hulk")
print(s)

# union
avengers = {"Iron Man", "Captain America"}
guardians = {"Star-Lord", "Rocket"}
all_heroes = s.union(guardians)
print(all_heroes)

# intersection
new_avengers = {"Spider-Man", "Doctor Strange", "Iron Man"}
common_heroes = s.intersection(new_avengers)
print(common_heroes)

# pop
print(s.pop())
print(s)

# update
s.update(["Black Widow", "Hawkeye"])
print(s)
```

```python
# Dictionary methods
marvel_movies = {
    "Iron Man": 2008,
    "The Avengers": 2012,
    "Guardians of the Galaxy": 2014
}

# keys
print(marvel_movies.keys())

# values
print(marvel_movies.values())

# items
print(marvel_movies.items())

# get
print(marvel_movies.get("Iron Man"))

# pop
print(marvel_movies.pop("The Avengers"))
print(marvel_movies)

# update
marvel_movies.update({"Black Panther": 2018, "Doctor Strange": 2016})
print(marvel_movies)
```
